```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Car Sale.csv')

df
```

```python
df.columns
```

```python
rename_map = {"Customer Name": "Customer_Name", "Annual Income": "Annual_Income", "Price ($)": "Price", "Dealer_No ": "Dealer_No", "Body Style": "Body_Style"}
df = df.rename(columns=rename_map)
```

```python
df.columns
```

```python
# 1. What is the average selling price of cars for each dealer, and how does it compare across different dealers?

def avg_price_by_dealer(df):
    return df.groupby('Dealer_Name')['Price'].mean().sort_values(ascending=False)

print("average selling price of cars for each dealer:")
print(avg_price_by_dealer(df))
```

```python
# 2. Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends?

def brand_highest_variation(data):
    var_by_brand = data.groupby("Company")["Price"].var()
    return var_by_brand.idxmax(), var_by_brand.max(), var_by_brand

print("highest variation in prices:")
print(brand_highest_variation(df))
```

```python
# 3. What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?

def price_distribution_by_transmission(data):
    desc = data.groupby("Transmission")["Price"].describe()
    iqr = desc["75%"] - desc["25%"]
    return desc, iqr

print("distribution of car prices for each transmission type:")
print(price_distribution_by_transmission(df))
```

```python
# 4. What is the distribution of car prices across different regions?

def price_distribution_by_region(df):
    return df.groupby("Dealer_Region")["Price"].describe()

print("distribution of car prices across different regions:")
print(price_distribution_by_region(df))
```

```python
# 5. What is the distribution of cars based on body styles?

def distribution_by_body_style(df):
    return df.groupby("Body_Style")["Price"].describe()

print("distribution of cars based on body styles:")
print(distribution_by_body_style(df))
```

```python
# 6. How does the average selling price of cars vary by customer gender and annual income?

def avg_price_by_gender_income(df):
    return df.groupby(["Gender", "Annual_Income"])["Price"].mean()

print("average selling price of cars vary by customer gender and annual income:")
print(avg_price_by_gender_income(df))
```

```python
# 7. What is the distribution of car prices by region, and how does the number of cars sold vary by region?

def price_distribution_by_region(df):
    return df.groupby("Dealer_Region")["Price"].describe()

print("distribution of car prices by region:")
print(price_distribution_by_region(df))
```

```python
# 8. How does the average car price differ between cars with different engine sizes?

def avg_price_by_engine(df):
    return df.groupby("Engine")["Price"].mean()

print("average car price differ between cars with different engine sizes:")
print(avg_price_by_engine(df))
```

```python
# 9. How do car prices vary based on the customer’s annual income bracket?

def price_distribution_by_income(df):
    return df.groupby("Annual_Income")["Price"].describe()

print("car prices vary based on the customer’s annual income bracket:")
print(price_distribution_by_income(df))
```

```python
# 10. What are the top 5 car models with the highest number of sales, and how does their price distribution look?

def top_5_models_by_sales(df):
    return df.groupby("Model")["Price"].count().nlargest(5)

print("top 5 car models with the highest number of sales:")
print(top_5_models_by_sales(df))
```

```python
# 11. How does car price vary with engine size across different car colors, and which colors have the highest price variation?

def price_by_engine_and_color(data):
    return data.groupby(["Engine", "Color"])["Price"].agg(["mean", "std", "count"])

print("car price vary with engine size across different car colors:")
print(price_by_engine_and_color(df))
```

```python
# 12.  Is there any seasonal trend in car sales based on the date of sale?

def seasonal_trend(data):
    temp = data.copy()
    temp["Date"] = pd.to_datetime(temp["Date"])
    temp["Month"] = temp["Date"].dt.month
    return temp["Month"].value_counts().sort_index()

print("Months:\n", seasonal_trend(df), "\n")
```

```python
# 13. How does the car price distribution change when considering different combinations of body style and transmission type?

def price_by_body_and_transmission(data):
    return data.groupby(["Body_Style", "Transmission"])["Price"].describe()

print("sample:\n", price_by_body_and_transmission(df).head(), "\n")
```

```python
# 14. What is the correlation between car price, engine size, and annual income of customers, and how do these features interact?

def corr_price_engine_income(data):
    temp = data.copy()
    temp["Engine_Code"] = temp["Engine"].astype("category").cat.codes
    return temp[["Price", "Annual_Income", "Engine_Code"]].corr()

print("Q14 corr:\n", corr_price_engine_income(df), "\n")
```

```python
# 15. How does the average car price vary across different car models and engine types?

def avg_price_model_engine(data):
    return data.pivot_table(index="Model", columns="Engine", values="Price", aggfunc="mean")

print("shape:", avg_price_model_engine(df).shape)
```
